```python
import pandas as pd
import numpy as np

# Sample schedule for the next 10 days
data = {
    'John':  [True, False, True,  False, True,  False, False, True, True,  False],
    'Judy':  [True, False, False, True,  True,  False, True,  True, False, False]
}
df = pd.DataFrame(data)
# Identify party days (i.e., when both John and Judy visit)
df['party'] = df['John'] & df['Judy']
print(len(df))
df
```

```python
# Re-create the 'party' column based on the logic used earlier
df['party'] = df['John'] & df['Judy']

# Initialize a list to hold the days_til_party values
days_til_party = [None] * len(df)
next_party_index = None

# Iterate in reverse order (from last day to first)
for i in reversed(range(len(df))):
    if df.loc[i, 'party']:
        next_party_index = i  # When a party is found, update the next party index
        days_til_party[i] = 0
    elif next_party_index is not None:
        days_til_party[i] = next_party_index - i
    else:
        days_til_party[i] = np.nan  # You can choose to set this to a specific value if there's no upcoming party

df.drop(columns=['party'], inplace=True)  # Drop the 'party' column if not needed
# Add the new column to the DataFrame
df['days_til_party'] = days_til_party
print(df)
```

```python
# Create a temporary column that holds the index on party days; otherwise, NaN
df['party_index'] = np.where(df['party'], df.index, np.nan)

# Reverse the Series, forward-fill, and then reverse back.
# This gives the index of the next party for every row.
df['next_party_index'] = df['party_index'][::-1].ffill()[::-1]

# Calculate the number of days until the next party by subtracting the current index from the next party index.
df['days_til_party'] = df['next_party_index'] - df.index

# Optionally, drop the temporary columns
df.drop(columns=['party', 'party_index', 'next_party_index'], inplace=True)
print(df)
```

```python
data = {
    'artist': ['A', 'B', 'A', 'A', 'B', 'C'],
    'venue': ['X', 'Y', 'X', 'Y', 'Y', 'Z'],
    'date': pd.to_datetime(['2024-01-10', '2024-01-15', '2024-02-20', 
                            '2024-01-22', '2024-02-10', '2024-02-11'])
}
df = pd.DataFrame(data)
print(df)
# count the number of concerts per (artist, venue), per year
# month. 
for i in range(len(df)):
    df.loc[i, 'year'] = df.loc[i, 'date'].year
    df.loc[i, 'month'] = df.loc[i, 'date'].month

for i in range(len(df)):
    df.loc[i, 'concerts'] = 1
# group by artist, venue, year, month and sum only numeric columns
df = df.groupby(['artist', 'venue', 'year', 'month'], as_index=False).agg({'concerts': 'sum'})
# rename the columns   
df.rename(columns={'concerts': 'number_of_concerts'}, inplace=True) 
print(df)
```
