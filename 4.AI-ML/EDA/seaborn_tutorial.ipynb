```python
# seaborn is a package for data visualization built over the matplotlib library
# importing packages
# if the package is not already installed, install them
# https://www.geeksforgeeks.org/python-seaborn-tutorial/
import seaborn as sns 

# loading dataset
data = sns.load_dataset("iris")

# # draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
# change in themes is best viewed if you comment this cell (otherwise some overlap may occur)
```

```python
# changing the theme
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# # loading dataset 
data = sns.load_dataset("iris") 

# # draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# # changing the theme to dark
sns.set_style("dark")
plt.show()
```

```python
# changing the dimensions of the figure
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# loading dataset 
data = sns.load_dataset("iris") 

# changing the figure size
plt.figure(figsize = (2, 4))

# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

# Removing the spines
sns.despine()

plt.show()
```

```python
# multiple subplots
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt 

# loading dataset 
data = sns.load_dataset("iris") 


def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data) 

with sns.axes_style('darkgrid'):
    
    # Adding the subplot
    plt.subplot(1, 2, 1)
    plot()
    
plt.subplot(1, 2, 2)
plot()
```

```python
# facet grid
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")
print(data)
# create grids based on the species column
plot = sns.FacetGrid(data, col="species")
plot.map(plt.plot, "sepal_width")

plt.show()
```

```python
# importing packages
# semantic mappings of subsets
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

# creating the relplot
# based on the specific species
sns.relplot(x='sepal_width', y='species', data=data)

plt.show()
```

```python
# scatter plot
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")
# x, y are column names
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

```python
# bar plot
# loading dataset
# can specify the x, y axes based on the column names
sns.barplot(x='species', y='sepal_length', data=data)
plt.show()
```

```python
# count plot
sns.countplot(x='species', data=data) # the number of samples are same in all the 3 species
plt.show()
```

```python
# box plot
sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
# violin plot
sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
# dist plot
sns.displot(data['sepal_width']) # in terms of histograms
plt.show()
```
