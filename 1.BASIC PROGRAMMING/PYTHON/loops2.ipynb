```python
from scipy.stats import binom

# Variables
n = 210
p = 0.95

# Cumulative probability for X <= 200
cumulative_prob = binom.cdf(200, n, p)

# Probability for X > 200
prob_over_200 = 1 - cumulative_prob

print("Probability that more than 200 passengers show up:", prob_over_200)
```

```python
from scipy.stats import poisson


# Mean (\(\mu\)) of the distribution
mu = 199.5

# Cumulative probability for X <= 200
cumulative_prob = poisson.cdf(200, mu)

# Probability for X > 200
prob_over_200 = 1 - cumulative_prob

print("Probability that more than 200 passengers show up:", prob_over_200)
```

```python
import numpy as np
from scipy.stats import gamma
import matplotlib.pyplot as plt

# Parameters for the Gamma distribution
k = 2.0  # shape parameter
theta = 2.0  # scale parameter

# Generate a range of x values
x = np.linspace(0, 20, 1000)

# Compute the CDF values
cdf_values = gamma.cdf(x, a=k, scale=theta)

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(x, cdf_values, label='Gamma CDF')
plt.xlabel('x')
plt.ylabel('CDF')
plt.title('Gamma Cumulative Distribution Function')
plt.legend()
plt.grid(True)
plt.show()
```
